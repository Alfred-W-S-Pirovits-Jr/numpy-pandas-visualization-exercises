```python
plt.rcParams["figure.figsize"] = [100, 100]
plt.rcParams["figure.autolayout"] = True

def f(x):
   return np.log(x) 

def g(x):
   return x 
x = np.linspace(-10, 10, 100)

def h(x):
   return np.log(x) * x

def i(x):
   return x ** 2

def j(x):
   return 2 ** x

# def k(x):
#     if x == 0:
#         break
#     else 

def l(x): 
    return x ** x
plt.plot(x, f(x), x, g(x), x, h(x), x, i(x), x, j(x), x, l(x), color='red')

plt.show()
```
