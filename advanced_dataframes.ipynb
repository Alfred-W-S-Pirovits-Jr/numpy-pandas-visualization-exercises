1.  Copy the users and roles DataFrames from the examples above.

```python
# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

2.  What is the result of using a right join on the DataFrames?

```python
users.merge(roles, left_on='role_id', right_on= 'id', how='left', indicator = True) 
```

It returns the appropriate information from roles discarding what is not needed

```python
users.merge(roles, left_on='role_id', right_on= 'id', how='right', indicator = True)
```

It maintains the fidelity of the roles table and adds appropriate info from users.  It discards the rest of the lines from users.

3.  What is the result of using an outer join on the DataFrames?

```python
users.merge(roles, left_on='role_id', right_on= 'id', how='outer', indicator = True) 
```

It gives us the 7 possibilities, the first six from users and the left over 7th possibility from the commenter that is unmatched.  Since admin author and reviewer appears on the users table it does not return those from roles.

4.  What happens if you drop the foreign keys from the DataFrames and try to merge them?

```python
users.merge(roles, how='outer', indicator = True) 
```

It gives us the 9 total possibilities by combining the two tables.  since we are not joining on id and role_id respectively it gives us all the possibilities

12.  Create a column named mileage_difference like you did in the DataFrames exercises; this column should contain the difference between highway and city mileage for each car.

```python
mpg_df = mpg_df.assign(mileage_difference = mpg_df.highway - mpg_df.city)
mpg_df
```

13.  Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.

```python
mpg_df = mpg_df.assign(average_mileage = (mpg_df.city + mpg_df.highway)/2)
mpg_df
```

14.  Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.

```python
mpg_df = mpg_df.assign(is_automatic = mpg_df.transmission.str.contains('auto'))
mpg_df
```
